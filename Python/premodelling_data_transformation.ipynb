```python
def transformation(dataframe, transformation_type, index, constant, column_name):
    ## ============== function explanation ==============
    ## This function performs lag, power and decay transformation
    ## ============== function parameter explanation ==============
    ## dataframe: input dataframe
    ## transformation.type: lag, decay or power
    ## transformation.type.index: index of transforamtion. i.e. lag1, lag2, decay1, decay2
    ## transformation.constant: the parameter used for transformation
    ## column.name: dataframe column needed for transformation
    
    if transformation_type.lower()=='lag':
        dataframe[['-'.join([column_name,transformation_type,str(index)])]] = dataframe[[column_name]].shift(periods=constant, fill_value=0)
        return
    elif transformation_type.lower()=='decay':
        new_col_name = '-'.join([column_name,transformation_type,str(index)])
        df.loc[0, new_col_name] = df.loc[0, column_name] * constant
        for i in range(1, len(df)):
            df.loc[i, new_col_name] = df.loc[i, column_name] * constant + df.loc[i-1, new_col_name] * (1-constant)
        return
    elif transformation_type.lower()=='pow':
        dataframe[['-'.join([column_name,transformation_type,str(index)])]] = dataframe[[column_name]].apply(lambda x: x**constant)
        return
    else:
        print('Only support lag, decay and power')
        return
```

```python
transformation(df, 'lag', 1, 1, 'National TV GRPs')
transformation(df, 'pow', 1, .8, 'National TV GRPs-lag-1')
transformation(df, 'decay', 1, .4, 'National TV GRPs-lag-1-pow-1')
```

```python
def transformation_sequence(dataframe, transformation_types, indexs, constants, column_name):
    ## ============== function explanation ==============
    ## This function performs a sequence of lag, power and decay transformation using transform
    ## ============== function parameter explanation ==============
    ## dataframe: input dataframe
    ## sequence: sequence of trasnformation
    ## sequence.indices: index number of the transformation
    ## values: value of transformation action
    ## column.name: name of column transformation operates on
    if len(transformation_types) != len(indexs) or len(indexs) != len(constants):
        print('not equal length')
        return
    for i in range(len(indexs)):
        transformation(dataframe, transformation_types[i], indexs[i], constants[i], column_name)
        column_name = '-'.join([column_name,transformation_types[i],str(indexs[i])])
        
    return
```

```python
transformation_sequence(df, ['lag','pow','decay'], [1,1,1], [1,.8,.4], 'National TV GRPs')
```
